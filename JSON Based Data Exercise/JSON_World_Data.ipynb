```python
import pandas as pd
import json
from pandas.io.json import json_normalize

#Loading JSON file into a pandas DataFrame#
world_bank = pd.read_json('world_bank_projects.json')

#Finding Top 10 Countries for Total Projects#
print('Exercise #1')
print(world_bank.countryname.value_counts().head(10))

#Finding Top 10 Major Project Themes#
with open('world_bank_projects.json') as file:
    data = json.load(file)
    result = json_normalize(data, 'mjtheme_namecode')
    print('\nExercise #2')
    print(result.name.value_counts().head(10))

#Adding Missing Names to Top 10 Project Themes#
    names_included = result.loc[result['name'] != ''].drop_duplicates()
    new = pd.merge(result, names_included, on=['code'], how='outer')
    new1 = new[['code', 'name_y']]
    print('\nExercise #3')
    print(new1.name_y.value_counts())

#Displaying Entire Dataframe with Missing Names Filled in#
    new1.columns = ['code', 'name']
    print('\n')
    print(new1)
```
